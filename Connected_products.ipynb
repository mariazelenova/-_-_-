```python
#проводим анализ ассоциаций, для этого смотрим параметры frequent item sets и association rules
```

```python
data_apr = out.groupby(["dt", "gtin"])[["cnt"]].sum().unstack().reset_index().fillna(0).set_index("dt")
```

```python
#data_apr.head()
```

```python
def num(x):
    if x <= 0:
        return 0
    elif x >=1:
        return 1
    
basket_new = data_apr.applymap(num)
```

```python
basket_new.head()
```

```python
#!pip install mlxtend
```

```python
def front(self, n):
    return self.iloc[:, :n]

def back(self, n):
    return self.iloc[:, -n:]

pd.DataFrame.front = front
pd.DataFrame.back = back
```

```python
#basket_new.head().front(10)
```

```python
from mlxtend.frequent_patterns import apriori

apr = apriori(basket_new.head(100).front(100), min_support = 0.02, use_colnames = True)
apr.sort_values(by = "support", ascending = False)
```

```python
#Это таблица выводится на выход с фразой "Выявлены связанные продукты."
#Должна быть возможность скачать файл
```

```python
from mlxtend.frequent_patterns import association_rules
end = association_rules(apr, metric = "lift", min_threshold = 1)
end = pd.DataFrame(end.sort_values(by = "confidence", ascending = False))
end.head()
```
